```python
df.drop(['CustomerID','Zip Code','Latitude','Longitude','Churn Label','Churn Score','Churn Reason'],axis=1,inplace=True)
```

```python
# First check which columns actually exist in your DataFrame
print(df.columns.tolist())

# Then drop only the columns that exist
# For example, if only some columns exist, you can use:
columns_to_drop = ['CustomerID', 'Zip Code', 'Latitude', 'Longitude', 'Churn Label', 'Churn Score', 'Churn Reason']
existing_columns = [col for col in columns_to_drop if col in df.columns]
df.drop(existing_columns, axis=1, inplace=True)

# Alternatively, you can use the errors='ignore' parameter to ignore non-existent columns
# df.drop(['CustomerID', 'Zip Code', 'Latitude', 'Longitude', 'Churn Label', 'Churn Score', 'Churn Reason'], 
#         axis=1, inplace=True, errors='ignore')
```

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
```

```python
df['Total Charges']
```

```python
df.isnull().sum()
```

```python
df['Total Charges'].fillna(df['Total Charges'].median(),inplace=True)
```

```python
x=df.drop('Churn Value',axis=1)
y=df['Churn Value']
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
cat_cols=x.select_dtypes(include='object').columns
```

```python
for col in cat_cols:
    x[col]=le.fit_transform(x[col])
```

```python
x.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1,stratify=y)
```

```python
ytrain.value_counts()
```

```python
ytest.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression(max_iter=1000,class_weight='balanced')
```

```python
log_model.fit(xtrain,ytrain)
```

```python
y_pred=log_model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('Accuracy',accuracy_score(ytest,y_pred))
```

```python
print(confusion_matrix(ytest,y_pred))
```

```python
print(classification_report(ytest,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=200,class_weight='balanced',random_state=1)
```

```python
rf.fit(xtrain,ytrain)
```

```python
rf_probs = rf.predict_proba(xtest)[:, 1]
rf_pred = (rf_probs >= 0.35).astype(int)
```

```python
print(classification_report(ytest, rf_pred))
```

```python
feature_importance = pd.DataFrame({
    'Feature': xtrain.columns,
    'Importance': rf.feature_importances_
}).sort_values(by='Importance', ascending=False)

feature_importance.head(10)
```
